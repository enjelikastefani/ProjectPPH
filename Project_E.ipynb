## Data cleaning

### Subtask:
Clean the data in the DataFrame `df`.


**Reasoning**:
Clean the data by handling missing values, removing duplicates, ensuring data type consistency, and standardizing the `tax_holiday_status` column.



```python
# Handle missing values
for col in df.select_dtypes(include=['number']):
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Remove duplicates
df.drop_duplicates(inplace=True)

# Ensure data type consistency
for col in df.select_dtypes(include=['object']):
    if col != 'tax_holiday_status' and col != 'tahun':
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        except:
            pass

# Standardize tax_holiday_status
df['tax_holiday_status'] = df['tax_holiday_status'].replace({'Ya': 1, 'Tidak': 0})
df = df[df['tax_holiday_status'].isin([0,1])]

display(df.head())
```

## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Calculate YoY changes, financial ratios, rolling averages, and analyze the impact of tax holidays on key financial metrics.



```python
# Year-over-Year Change Calculation
for col in ['pendapatan_penjualan', 'beban_operasional', 'laba_kotor_penjualan']:
    df[f'{col}_yoy'] = df[col].pct_change() * 100

# Financial Ratios
df['gross_profit_margin'] = df['laba_kotor_penjualan'] / df['pendapatan_penjualan']
# Handle division by zero
df['gross_profit_margin'] = df['gross_profit_margin'].replace([np.inf, -np.inf], np.nan)
df['operating_profit_margin'] = (df['laba_kotor_operasional'] - df['beban_operasional']) / df['pendapatan_penjualan']
# Handle division by zero
df['operating_profit_margin'] = df['operating_profit_margin'].replace([np.inf, -np.inf], np.nan)


# Rolling Averages
for window in [2, 3]:
    df[f'pendapatan_penjualan_rolling_{window}yr'] = (
        df['pendapatan_penjualan'].rolling(window=window, min_periods=1).mean()
    )

# Tax Holiday Impact Analysis
# Calculate the change in pendapatan_penjualan in the year following a tax holiday year
df['pendapatan_penjualan_post_tax_holiday'] = 0
for i in range(1, len(df)):
    if df['tax_holiday_status'].iloc[i - 1] == 1:
        df['pendapatan_penjualan_post_tax_holiday'].iloc[i] = df['pendapatan_penjualan'].iloc[i] - df['pendapatan_penjualan'].iloc[i - 1]


display(df.head())
```

## Model training

### Subtask:
Train a time series model to forecast key financial metrics.


**Reasoning**:
Train a time series model to forecast key financial metrics, starting with a simple model like ARIMA.



```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Prepare the data
df = df.sort_values('tahun')  # Ensure data is sorted by time
target_metric = 'pendapatan_penjualan'
data = df[target_metric].values
train_data = data[:-1]  # Use all but the last year for training
test_data = data[-1:]    # Use the last year for testing

# Train the ARIMA model
try:
    # Find optimal p, d, q values (simplified example)
    p, d, q = (5,1,0) # these values may not be optimal, need more investigation
    model = ARIMA(train_data, order=(p, d, q))
    model_fit = model.fit()

    # Make predictions
    predictions = model_fit.predict(start=len(train_data), end=len(data)-1)

    # Evaluate the model
    rmse = np.sqrt(mean_squared_error(test_data, predictions))
    print(f"RMSE: {rmse}")

    # Visualize results
    plt.figure(figsize=(10, 6))
    plt.plot(df['tahun'], data, label='Actual')
    plt.plot(df['tahun'][-len(predictions):], predictions, label='Predicted')
    plt.xlabel('Year')
    plt.ylabel(target_metric)
    plt.title('ARIMA Forecast')
    plt.legend()
    plt.show()

except Exception as e:
    print(f"An error occurred during model training or prediction: {e}")
```

```python
from prophet import Prophet
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Prepare data for Prophet
prophet_df = pd.DataFrame({'ds': pd.to_datetime(df['tahun']), 'y': df['pendapatan_penjualan']})

# Split the data
train_prophet = prophet_df[:-1]
test_prophet = prophet_df[-1:]

# Train the Prophet model
model = Prophet()
try:
    model.fit(train_prophet)
    future = model.make_future_dataframe(periods=1, freq='Y')
    forecast = model.predict(future)

    # Evaluate the model
    y_true = test_prophet['y'].values
    y_pred = forecast['yhat'][-1:]
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"RMSE for Prophet: {rmse}")

    # Visualize the forecast
    fig1 = model.plot(forecast)
    plt.show()

    fig2 = model.plot_components(forecast)
    plt.show()
except Exception as e:
    print(f"An error occurred: {e}")
```
