## Data loading

### Subtask:
Load the "combined_financial_data.csv" file into a pandas DataFrame.


**Reasoning**:

```python
import pandas as pd

try:
    df = pd.read_csv('combined_financial_data.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'combined_financial_data.csv' not found.")
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file. Please check its format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the loaded financial data to understand its structure, identify potential issues, and gain initial insights.


**Reasoning**:
Examine the data structure, identify missing values, analyze data types, and explore key variables as instructed in the subtask.

## Data cleaning

### Subtask:
Clean the data by handling missing values and outliers.


**Reasoning**:
Impute missing values and handle outliers for the specified columns.



```python
# Impute missing values
for col in ['aset_id', 'aset_kategori', 'aset_nilai_perolehan', 'aset_umur_ekonomis', 'aset_metode', 'aset_tahun_perolehan']:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            # Impute with the median for numerical columns
            df[col] = df[col].fillna(df[col].median())
        else:
            # Impute with the mode for categorical columns
            df[col] = df[col].fillna(df[col].mode()[0])

# Handle outliers using IQR
numerical_cols = ['transaksi_pendapatan', 'transaksi_beban_operasional', 'transaksi_penyusutan', 'transaksi_laba_kena_pajak', 'aset_nilai_perolehan', 'aset_umur_ekonomis', 'aset_tahun_perolehan']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# Visualize the distributions after outlier handling
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True, color=f"C{i}")
    plt.title(f'Distribution of {col} after outlier handling')

plt.tight_layout()
plt.show()

display(df.head())
```

## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Convert the 'tahun' column to datetime objects, one-hot encode categorical features, create new features (profit margin), and remove irrelevant columns.



```python
import pandas as pd

# 1. Convert 'tahun' to datetime (if needed)
if 'tahun' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['tahun']):
    try:
        df['tahun'] = pd.to_datetime(df['tahun'], format='%Y')
    except ValueError:
        print("Warning: Could not convert 'tahun' column to datetime.  It might not be in a recognizable format.")


# 2. One-hot encode categorical features
categorical_cols = ['aset_metode', 'aset_kategori', 'transaksi_skenario']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)


# 3. Create new features
df['profit_margin'] = df['transaksi_laba_kena_pajak'] / df['transaksi_pendapatan']
df['depreciation_rate'] = df['transaksi_penyusutan'] / df['aset_nilai_perolehan']


# 4. Remove irrelevant columns (if any)
# Example: Remove 'kebijakan_tax_holiday_awal' and 'kebijakan_tax_holiday_akhir' if deemed unnecessary
# The reasoning for removing these columns is that they are not directly related to the financial performance.
if 'kebijakan_tax_holiday_awal' in df.columns:
    df = df.drop(columns=['kebijakan_tax_holiday_awal', 'kebijakan_tax_holiday_akhir'])


display(df.head())
```

## Data analysis

### Subtask:
Perform a comprehensive analysis of the wrangled financial data.


**Reasoning**:
Calculate descriptive statistics, analyze correlations, and perform time series analysis on the financial data.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
display(df.describe())
print("\nDescriptive Statistics for 'profit_margin' and 'depreciation_rate':")
display(df[['profit_margin', 'depreciation_rate']].describe())


# 2. Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Financial Variables')
plt.show()

# Identify strong correlations
strong_correlations = correlation_matrix[abs(correlation_matrix) > 0.7]
print("\nStrong Correlations (absolute value > 0.7):\n", strong_correlations)


# 3. Time Series Analysis (assuming 'tahun' is a time series)
# Convert 'tahun' to datetime if not already
if not pd.api.types.is_datetime64_any_dtype(df['tahun']):
    df['tahun'] = pd.to_datetime(df['tahun'])


plt.figure(figsize=(14, 6))
sns.lineplot(x='tahun', y='transaksi_pendapatan', data=df, marker='o', label='Revenue')
sns.lineplot(x='tahun', y='transaksi_beban_operasional', data=df, marker='x', label='Expenses')
sns.lineplot(x='tahun', y='transaksi_laba_kena_pajak', data=df, marker='s', label='Profit')
plt.title('Financial Performance Over Time')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.show()

# Calculate year-over-year changes (example for revenue)
df['revenue_yoy_change'] = df['transaksi_pendapatan'].pct_change() * 100
print("\nYear-over-year Revenue Change:\n", df[['tahun', 'revenue_yoy_change']])
```
